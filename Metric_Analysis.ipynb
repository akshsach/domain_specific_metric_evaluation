```python
import numpy as np
from scipy.stats import friedmanchisquare

finance_data = {
    "F1-Score": [0.92, 0.88, 0.85],
    "AUC-ROC": [0.93, 0.90, 0.86],
    "Precision": [0.91, 0.87, 0.84],
    "Recall": [0.89, 0.86, 0.83],
    "MCC": [0.87, 0.85, 0.82]}

finance_matrix = np.array(list(finance_data.values()))
statistic, p_value = friedmanchisquare(*finance_matrix)

print("Friedman Test Results for Finance Domain:")
print("F-statistic:", statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("Significant difference between metrics; perform post-hoc analysis.")
else:
    print("No significant difference between metrics.")
```

```python
import numpy as np
from scipy.stats import rankdata

metric_names = list(finance_data.keys())
finance_matrix = np.array(list(finance_data.values()))

ranks = np.array([rankdata(-scores) for scores in finance_matrix.T])
mean_ranks = np.mean(ranks, axis=0)

k = finance_matrix.shape[0]
N = finance_matrix.shape[1]

from scipy.stats import t
q_alpha = 2.569  # For alpha=0.05 and large N
CD = q_alpha * np.sqrt((k * (k + 1)) / (6.0 * N))

print("Nemenyi Post-Hoc Test Results")
print("Critical Difference (CD):", CD)
for i in range(k):
    for j in range(i + 1, k):
        diff = abs(mean_ranks[i] - mean_ranks[j])
        if diff > CD:
            print(f"Significant difference between {metric_names[i]} and {metric_names[j]} (Diff: {diff:.3f})")
        else:
            print(f"No significant difference between {metric_names[i]} and {metric_names[j]} (Diff: {diff:.3f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rankdata

metrics = ["F1-Score", "AUC-ROC", "Precision", "Recall", "MCC"]
mean_performance = [0.92, 0.93, 0.91, 0.89, 0.87]

ranks = rankdata([-x for x in mean_performance])

inverse_ranks = 1 / ranks
proportions = inverse_ranks / sum(inverse_ranks)

plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=metrics, autopct="%1.1f%%", startangle=140,
        colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#c2c2f0'])
plt.title("Suitability of Evaluation Metrics for the Finance Domain")
plt.show()
```
